```python
import matplotlib.pyplot as plt
from dataset import SensorDataset
```

```python
in_columns = [
    # "ACC_x",
    # "ACC_y",
    # "ACC_z",
    # "GRVT_x",
    # "GRVT_y",
    # "GRVT_z",
    "MAG_x",
    "MAG_y",
    "MAG_z",
    # "ORT_a_cos",
    # "ORT_a_sin",
    # "ORT_p_cos",
    # "ORT_p_sin",
    # "ORT_r_cos",
    # "ORT_r_sin",
]
mode = "train"
dataset = SensorDataset(
    in_columns=in_columns, mode=mode, data_dir="../data", data_fname="da23_sensor_data.npz"
)
```

```python
class_1_ids = [i for i, j in enumerate(dataset.y.squeeze().tolist()) if j == 1.0]
class_1_ids
```

```python
print(f'dataset length : {len(dataset)}')
print(f'x.shape : {dataset.x.shape}')
print(f'y.shape : {dataset.y.shape}')
sample_idx = 0
sample_x = dataset.x[sample_idx].T.numpy()
sample_x = (sample_x - sample_x.mean(axis=0)) / sample_x.std(axis=0)  # normalize
sample_y = 'pos' if dataset.y[sample_idx].numpy() else 'neg'
plt.plot(sample_x)
plt.title(f'{mode} data : {sample_idx}-th sample : y={sample_y}')
plt.legend(in_columns)
plt.show()
```
